```python
"""
1. Check dataset
"""
from sklearn.datasets import load_wine

# 1.1 Load data
wine = load_wine()

# 1.2. Check keys
wine.keys()
```

```python
# 1.2. Assign data : Feature Data, Label Data 지정하기
wine_data = wine.data
print(wine_data)

wine_label = wine.target
print(wine_label)
```

```python
# 1.4. Check columns : Target Names 출력해 보기
print(wine.feature_names)
print(wine.target_names)
```

```python
# 1.5. Check description
print(wine.DESCR)
```

```python
"""
2. Set train, validation data
"""
import pandas as pd

# 2.1. Make DataFrame
wine_df = pd.DataFrame(data = wine_data, columns = wine.feature_names)
wine_df.head()
```

```python
# 2.2. Add Label column
wine_df['label'] = wine_label
wine_df.head()
```

```python
# 2.3. Split train, test data : train, test 데이터 분리
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(wine_data,
                                                    wine_label,
                                                    test_size = 0.2,
                                                    random_state = 7
                                                    )
print(f'wine_data 개수: {len(wine_data)}, X_train 개수: {len(x_train)}, X_test 개수: {len(x_test)}')
print(f'x_train.shape: {x_train.shape}, x_test.shape: {x_test.shape}')
print(f'y_train.shape: {y_train.shape}, y_test.shape: {y_test.shape}')
```

```python
"""
3. Model select and train
"""
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

# 3.1. Assign model
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'SVM': svm.SVC(),
    'SGD': SGDClassifier(),
    'Logistic Regression': LogisticRegression(max_iter = 10000)
}

# Result
results = {}
```

```python
# 3.2. Model train, predict
for name, model in models.items():
  curr_model = models[name]
  curr_model.fit(x_train, y_train)
  y_pred = curr_model.predict(x_test)
  accuracy = accuracy_score(y_test, y_pred)
  results[name] = {
      'accuracy': accuracy,
      'predictions': y_pred
  }
```

```python
# 3.3. Model comparison
for name, result in results.items():
    print(f"\n{name}:")
    print(f"Accuracy: {result['accuracy']}")
    print("Classification Report:")
    print(classification_report(y_test, result['predictions']))
```
